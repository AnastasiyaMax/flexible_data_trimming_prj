Создадим список из 23 датасетов, которые нужно преобразовать(GSE22358, GSE76360 и TCGA-UC не учитываем, так как в них количества ответчиков и неответчиков несбалансированы):

```
data = ['GSE18728.csv',  'GSE25066.csv',  'GSE50948.csv',  'GSE9782.csv',	'TCGA_LGG.csv',
        'GSE20181.csv',  'GSE32646.csv',  'GSE5122.csv',   'TARGET10.csv', 'GSE20194.csv',
        'GSE37946.csv',  'GSE55145.csv',  'TARGET20Busulfan.csv','GSE20271.csv', 
        'GSE39754.csv',  'GSE59515.csv',  'TARGET20NoBusulfan.csv',
        'GSE41998.csv',  'GSE68871.csv',  'TARGET50.csv',
        'GSE23988.csv',  'GSE42822.csv',  'TCGA_LC.csv']
```

Преобразовываем все датасеты к нужному виду:

```
for i in data:
  df = pd.read_csv('/content/drive/My Drive/FlowPS/data/' + i)
  df2 = df.T
  df2 = df2.reset_index(drop=True)

  new_header = df2.iloc[0] 
  df2 = df2[1:] 
  df2.columns = new_header 
  df2 = df2.drop(df2.columns[[0]], axis='columns')

  df2['Sample ID'] = df2.index
  cols = df2.columns.tolist()
  cols = cols[-1:] + cols[:-1]
  df2 = df2[cols]

  df2.to_csv('/content/drive/My Drive/FlowPS/change_data/' +  i, index=False)
```
